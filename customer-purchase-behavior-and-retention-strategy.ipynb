# Bar Chart: Purchase Frequency by Region

# Sankey Diagram: Retention Strategies to Preferred Purchase Times


```python

import plotly.graph_objects as go

# Define nodes and links for Sankey Diagram
nodes = list(set(df["Retention_Strategy"]) | set(df["Preferred_Purchase_Times"]))
node_indices = {node: idx for idx, node in enumerate(nodes)}
links = {
    "source": df["Retention_Strategy"].map(node_indices),
    "target": df["Preferred_Purchase_Times"].map(node_indices),
    "value": df["Purchase_Frequency"],
}

sankey_fig = go.Figure(
    go.Sankey(
        node=dict(label=nodes),
        link=dict(
            source=links["source"],
            target=links["target"],
            value=links["value"],
        ),
    )
)
sankey_fig.update_layout(title_text="Flow of Retention Strategies to Preferred Purchase Times", font_size=10)
sankey_fig.show(renderer='iframe_connected')
```
